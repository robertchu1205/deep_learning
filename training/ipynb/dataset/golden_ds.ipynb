```python
import sqlite3
from shutil import copyfile
import os
import csv
db_path = '/p3/metadata.db'
conn = sqlite3.connect(db_path)
c = conn.cursor()
```

# model1

## folder

```python
tar_path = '/p3/golden-ds/alucap-m1-folder/'
m1_class = ['OK', 'NG-MoreComp', 'NG-OutsidePosition', 'NG-UpsideDown', 'NG-NoneComp']
com = 'AluCap'
for m1 in m1_class:
    count = 0
    if m1 == 'NG-MoreComp':
        label = 'NG-UpsideDown'
    else:
        label = m1
    if not os.path.exists(os.path.join(tar_path, label)):
        os.makedirs(os.path.join(tar_path, label))
    symbol = (com, m1, )
    dates = c.execute('''
                select distinct date from metadata
                where extension = 'png'
                and component = ?
                and test_label = ?
            ''', symbol).fetchall()
    print(f'{m1}, date', len(dates))
    su = 0
    for d in dates:
        symbol = (com, m1, d[0],)
        locations = c.execute('''
                select distinct location from metadata
                where extension = 'png'
                and component = ?
                and test_label = ?
                and date = ?
            ''', symbol).fetchall()
#         print('loc', len(locations))
        su += len(locations)
        for l in locations:
            symbol = (com, m1, d[0], l[0])
            if label == 'OK':
                x = c.execute('''
                    select * from metadata
                    where extension = 'png'
                    and component = ?
                    and test_label = ?
                    and date = ?
                    and location = ?
                    ORDER BY RANDOM() LIMIT 10
                ''', symbol).fetchall()
            else:
                x = c.execute('''
                    select * from metadata
                    where extension = 'png'
                    and component = ?
                    and test_label = ?
                    and date = ?
                    and location = ?
                ''', symbol).fetchall()
            count+=len(x)
            for xx in x:
#                 print(xx[0], os.path.join(tar_path, label, xx[1]))
                copyfile(xx[0], os.path.join(tar_path, label, xx[1]))
#     print('avg location', su/len(dates))
    print('su', su)
    print(m1, count)
```

```python
m1_class = ['NG-OutsidePosition', 'NG-NoneComp'] 
tar_path = '/p3/golden-ds/alucap-m1'
with open(os.path.join(tar_path, 'OK.tsv'), 'r') as out_file:
    rows = csv.reader(out_file)
    rows = list(rows)
    print(rows[0][0].split('\t'))
    ok = len(rows)
    print('ok', ok)
    
for m1 in m1_class:
    with open(os.path.join(tar_path, f'{m1}-V2.tsv'), 'wt') as out_file:
        tsv_writer = csv.writer(out_file, delimiter='\t')
        with open(os.path.join(tar_path, f'{m1}.tsv'), 'r') as read_file:
            rows = csv.reader(read_file)
            rows = list(rows)
            print('ori', len(rows))
            new_row = 0
            pointer = 0
            while new_row!=ok:
                if pointer==len(rows):
                    print('pointer to 0')
                    pointer=0
                splitted = rows[pointer][0].split('\t')
                tsv_writer.writerow([splitted[0], splitted[1], splitted[2]])
                new_row+=1
                pointer+=1
    print(new_row)
```

```python
#'NG-UpsideDown', 'NG-MoreComp'
with open(os.path.join(tar_path, f'NG-UpsideDown-V2.tsv'), 'wt') as out_file:
    tsv_writer = csv.writer(out_file, delimiter='\t')
    with open(os.path.join(tar_path, f'NG-UpsideDown.tsv'), 'r') as read_file:
        rows = csv.reader(read_file)
        u_rows = list(rows)
        print('ori_UD', len(u_rows))
        with open(os.path.join(tar_path, f'NG-MoreComp.tsv'), 'r') as read_file_2:
            rows = csv.reader(read_file_2)
            m_rows = list(rows)
            print('ori_MC', len(m_rows))
            all_rows = u_rows+m_rows
            print('ori_two_types', len(all_rows))

            new_row = 0
            pointer = 0
            while new_row!=ok:
                if pointer==len(all_rows):
                    print('pointer to 0')
                    pointer=0
                splitted = all_rows[pointer][0].split('\t')
                tsv_writer.writerow([splitted[0], splitted[1], splitted[2]])
                new_row+=1
                pointer+=1
print(ok)
```

```python
m2_class = ['0', '180', '270', '90']
tar_path = '/p3/golden-ds/eleccap-degree-folder'
com = 'ElecCap'
for m2 in m2_class:
    if m2 == '180' or m2 == '90':
        label = 'NG-InversePolarity'
        degree = str((int(m2)+180)%360)
    else:
        label = 'OK'
        degree = m2
    if not os.path.exists(os.path.join(tar_path, m2)):
        os.makedirs(os.path.join(tar_path, m2))
    print(label, degree)
    count = 0
    symbol = (com, label, degree,)
    dates = c.execute('''
                select distinct date from metadata
                where extension = 'png'
                and component = ?
                and test_label = ?
                and degree = ?
            ''', symbol).fetchall()
    print(f'{m2}, date', len(dates))
    su = 0
    for d in dates:
        symbol = (com, label, degree, d[0])
        locations = c.execute('''
                select distinct location from metadata
                where extension = 'png'
                and component = ?
                and test_label = ?
                and degree = ?
                and date = ?
            ''', symbol).fetchall()
        su += len(locations)
        for l in locations:
            symbol = (com, label, degree, d[0], l[0])
            if label=='OK':
                x = c.execute('''
                    select * from metadata
                    where extension = 'png'
                    and component = ?
                    and test_label = ?
                    and degree = ?
                    and date = ?
                    and location = ?
                    ORDER BY RANDOM() LIMIT 10
                ''', symbol).fetchall()
            else:
                x = c.execute('''
                    select * from metadata
                    where extension = 'png'
                    and component = ?
                    and test_label = ?
                    and degree = ?
                    and date = ?
                    and location = ?
                ''', symbol).fetchall()
            count+=len(x)
            for xx in x:
                copyfile(xx[0], os.path.join(tar_path, m2, xx[1]))
#     print('avg location', su/len(dates))
    print('su', su)
    print(m2, count)
```
